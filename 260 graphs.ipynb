```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.rc('axes', titlesize=16)
plt.rc('axes', labelsize=14)
plt.rc('xtick', labelsize=12)
plt.rc('ytick', labelsize=12)

```

```python
df = pd.read_csv("knockpy_times.csv")
```

```python
df
```

```python
df_ram = df.groupby('RAM').mean()
df_ram.reset_index(inplace=True)
sns.set_style("whitegrid")
sns.scatterplot(x='RAM', y='time', data=df_ram).set_title('Average time vs RAM')
```

```python
df_1 = df[df['function'] == 'dot']
sns.set_style("whitegrid")
fig, ax = plt.subplots()
fig.set_size_inches(8, 6)
sns.barplot(x='RAM', y='time', hue='num_cores', data=df_1).set_title('Dot function time by RAM and Number of Cores')
ax.set_xlabel("RAM (GB)")
ax.set_ylabel("time (s)")

```

```python
df_1 = df[df['function'] == 'eigh']
sns.set_style("whitegrid")
fig, ax = plt.subplots()
fig.set_size_inches(8, 6)
sns.barplot(x='RAM', y='time', hue='num_cores', data=df_1).set_title('Eigh function time by RAM and Number of Cores')
ax.set_xlabel("RAM (GB)")
ax.set_ylabel("time (s)")
```

```python
df_1 = df[df['function'] == 'transpose']
sns.set_style("whitegrid")
fig, ax = plt.subplots()
fig.set_size_inches(8, 6)
sns.barplot(x='RAM', y='time', hue='num_cores', data=df_1).set_title('Transpose function time by RAM and Number of Cores')
ax.set_xlabel("RAM (GB)")
ax.set_ylabel("time (s)")
```

```python
df_1 = df[df['function'] == 'inv']
sns.set_style("whitegrid")
fig, ax = plt.subplots()
fig.set_size_inches(8, 6)
sns.barplot(x='RAM', y='time', hue='num_cores', data=df_1).set_title('Inverse function time by RAM and Number of Cores')
ax.set_xlabel("RAM (GB)")
ax.set_ylabel("time (s)")
```

```python
df_1 = df[(df['RAM'] == 15)]
sns.set_style("whitegrid")
fig, ax = plt.subplots()
fig.set_size_inches(6, 4)
plot = sns.scatterplot(x='num_cores', y='time', hue='function', style='chunk_type', data=df_1, ax=ax)
plot.set(xlabel="Number of Cores", title="Time by Number of Cores for RAM=15GB", xticks=[1, 4, 8])
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

```

```python
df_1 = df[(df['RAM'] == 15) & (df['function'] == 'dot')]
sns.set_style("whitegrid")
fig, ax = plt.subplots()
fig.set_size_inches(6, 4)
plot = sns.scatterplot(x='num_cores', y='time', color='blue', style='chunk_type', data=df_1, ax=ax)
plot.set(xlabel="Number of Cores", title="Time by Number of Cores for RAM=15GB and Dot function", xticks=[1, 4, 8])
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

```

```python
df_1 = df[(df['RAM'] == 15) & (df['function'] == 'eigh')]
sns.set_style("whitegrid")
fig, ax = plt.subplots()
fig.set_size_inches(6, 4)
plot = sns.scatterplot(x='num_cores', y='time', color='orange', style='chunk_type', data=df_1, ax=ax)
plot.set(xlabel="Number of Cores", title="Time by Number of Cores for RAM=15GB and Eigh function", xticks=[1, 4, 8])
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
df_1 = df[(df['RAM'] == 15) & (df['function'] == 'transpose')]
sns.set_style("whitegrid")
fig, ax = plt.subplots()
fig.set_size_inches(6, 4)
plot = sns.scatterplot(x='num_cores', y='time', color='green', style='chunk_type', data=df_1, ax=ax)
plot.set(xlabel="Number of Cores", title="Time by Number of Cores for RAM=15GB and Transpose function", xticks=[1, 4, 8])
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

```
